```python
indep_var=['GapPct_25_', 'GapPct_51_', 'GapPct_101',
       'GapPct_200', 'GapPct_251', 'Herbaceous_h', 'Woody_h', 'BareSoilCo',
       'TotalFolia', 'SageBrushC', 'SagebrushH', 'NonInvPereF', 'NonInvPeG',
       'NonInvAnnG', 'NonInvAnnF', 'NonInvShru', 'InvAnnGras', 'InvAnnForb']
```

```python
dep_data=reindex_raw.drop(indep_var,axis=1)
```

```python
dep_data.drop(['lat','lon'],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor     
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
r2=[]
mse=[]
rmse=[]
fea_import=[]
oob=[]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(dep_data, reindex_raw[indep_var], test_size=0.2, random_state=0)
```

```python
pred2016=pd.read_csv(r'data/prediction/input/pred2016_input.txt')
```

```python
pred2016.drop(['FID','Field19','Field20'],axis=1,inplace=True)
pred2016.head()
```

```python
for col in indep_var:
    rf_reg=RandomForestRegressor(oob_score=True,random_state=0,n_jobs=-1)   
    rf_reg.fit(X_train,Y_train[col])
    pred_Y=rf_reg.predict(X_test)
    #prediction
    output=rf_reg.predict(pred2016)
    np.savetxt(r"pred2016/temp/single_pred/"+col+'.txt',output,delimiter=',',fmt='%10.5f')
    #output metrics
    r2.append(r2_score(Y_test[col],pred_Y))
    mse.append(mean_squared_error(Y_test[col],pred_Y))
    rmse.append(sqrt(mean_squared_error(Y_test[col],pred_Y)))
    fea_import.append(rf_reg.feature_importances_)
    oob.append(rf_reg.oob_score_)
```

```python
np.savetxt(r'data\regression\metrics\single\r_square_single.txt',r2,delimiter=',',fmt='%10.5f')
np.savetxt(r'data\regression\metrics\single\rmse_single.txt',rmse,delimiter=',',fmt='%10.5f')    
np.savetxt(r'data\regression\metrics\single\mse_single.txt',mse,delimiter=',',fmt='%10.5f')  
np.savetxt(r'data\regression\metrics\single\fea_import_single.txt',fea_import,delimiter=',',fmt='%10.5f')
np.savetxt(r'data\regression\metrics\single\oob.txt_single.txt',oob,delimiter=',',fmt='%10.5f')
```
